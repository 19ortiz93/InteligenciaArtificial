```python
#Número de clusters y creacion de clusters
m = 2

V1 = np.zeros((2,1))
V2 = np.zeros((2,1))


V1[0,0]=np.random.random()
V1[1,0]=np.random.random()
V2[0,0]=np.random.random()
V2[1,0]=np.random.random()

```

```python
#Calcular funcion de pertenencia
def pertenencia():
  # = clusters()

  MC1=0
  MC2=0

  MCx=np.zeros((11,2))
  #np.sum((data[0,0]))
  for i in range(11):
  #for j in range(2):
    x1v1 = (data[i,0]-V1[0,0])**2+(data[i,1]-V1[1,0])**2
    x1v2 = (data[i,0]-V2[0,0])**2+(data[i,1]-V2[1,0])**2
    MC1=(1/((x1v1/x1v1)+(x1v1/x1v2)))
    MC2=(1/((x1v2/x1v1)+(x1v2/x1v2)))
    MCx[i,0] = MC1
    MCx[i,1] = MC2

  #print(MCx)
  return MCx
```

```python
#actualizar v1 y v2
def actualizarV():
  MCx = pertenencia()

  #VA1 = np.zeros((2,1))
  #VA1[1,0] = np.sum(np.multiply((MCx[:,0])**2,data[:,1]))/np.sum(MCx[:,0]**2)
  #VA1[0,0] = np.sum(np.multiply((MCx[:,0])**2,data[:,0]))/np.sum(MCx[:,0]**2)

  #V1 = np.zeros((2,1))
  V1[1,0] = np.sum(np.multiply((MCx[:,0])**2,data[:,1]))/np.sum(MCx[:,0]**2)
  V1[0,0] = np.sum(np.multiply((MCx[:,0])**2,data[:,0]))/np.sum(MCx[:,0]**2)

  #V2 = np.zeros((2,1))
  V2[1,0] = np.sum(np.multiply((MCx[:,1])**2,data[:,1]))/np.sum(MCx[:,1]**2)
  V2[0,0] = np.sum(np.multiply((MCx[:,1])**2,data[:,0]))/np.sum(MCx[:,1]**2)

  #print(VA1)
  #print(VA2)
  return V1, V2, MCx
```
